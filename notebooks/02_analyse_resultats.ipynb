# 02 - Analyse des rÃ©sultats

Analyse locale du fichier de sortie : scores, statuts, catÃ©gories, types de dÃ©tection et priorisation.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_ROOT = Path("..").resolve()
OUTPUT_CSV = PROJECT_ROOT / "output" / "resultats_controle.csv"

df = pd.read_csv(OUTPUT_CSV, encoding="utf-8-sig")
df.head()
```

## 1. SynthÃ¨se par document

```python
doc_scores = (
    df.groupby("document", as_index=False)
      .agg(score=("score", "max"), statut=("statut", "first"), nb_alertes=("terme_detecte", lambda s: (s != "").sum()))
      .sort_values(["score", "nb_alertes"], ascending=[False, False])
)
display(doc_scores)
print("Nombre de documents :", len(doc_scores))
print("Score moyen :", round(doc_scores["score"].mean(), 2))
```

## 2. Documents prioritaires

```python
prioritaires = doc_scores[doc_scores["score"] >= 80]
display(prioritaires)
```

## 3. RÃ©partition des statuts

```python
statuts = doc_scores["statut"].value_counts().reset_index()
statuts.columns = ["statut", "nombre_documents"]
display(statuts)

plt.figure(figsize=(8, 4))
plt.bar(statuts["statut"], statuts["nombre_documents"])
plt.xticks(rotation=45, ha="right")
plt.title("RÃ©partition des documents par statut")
plt.tight_layout()
plt.show()
```

## 4. Analyse par type de dÃ©tection

```python
df_alertes = df[df["score"] > 0].copy()
types = df_alertes["type_detection"].value_counts().reset_index()
types.columns = ["type_detection", "nombre_alertes"]
display(types)

if not types.empty:
    plt.figure(figsize=(8, 4))
    plt.bar(types["type_detection"], types["nombre_alertes"])
    plt.xticks(rotation=45, ha="right")
    plt.title("Alertes par type de dÃ©tection")
    plt.tight_layout()
    plt.show()
```

## 5. Analyse par catÃ©gorie Article 9

```python
categories = df_alertes["categorie"].value_counts().reset_index()
categories.columns = ["categorie", "nombre_alertes"]
display(categories)
```

## 6. Croisement catÃ©gorie / type de dÃ©tection

```python
pivot = pd.pivot_table(
    df_alertes,
    index="categorie",
    columns="type_detection",
    values="document",
    aggfunc="count",
    fill_value=0
)
display(pivot)
```

## 7. Top termes dÃ©clencheurs

```python
top_termes = df_alertes["terme_detecte"].value_counts().reset_index()
top_termes.columns = ["terme_detecte", "nombre_occurrences"]
display(top_termes.head(30))
```

## 8. DÃ©tail des alertes fortes

```python
alertes_fortes = df_alertes[df_alertes["score"] >= 80].sort_values(["score", "document"], ascending=[False, True])
display(alertes_fortes[["document", "score", "score_unitaire", "statut", "terme_detecte", "terme_reference", "type_detection", "categorie", "motif", "page", "contexte"]])
```

## 9. Export synthÃ¨se

```python
synthese_file = PROJECT_ROOT / "output" / "synthese_documents.csv"
doc_scores.to_csv(synthese_file, index=False, encoding="utf-8-sig")
print("SynthÃ¨se gÃ©nÃ©rÃ©e :", synthese_file)
```
